```python
# TASK 3: Calculate the distances between a launch site to its proximities
import math

def haversine(coord1, coord2):
    # Coordinates in decimal degrees (e.g. 28.5623, -80.5774)
    lat1, lon1 = coord1
    lat2, lon2 = coord2

    # Convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1 
    dlon = lon2 - lon1 
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a)) 

    # Radius of Earth in kilometers (mean radius)
    r = 6371  
    return c * r
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
import folium
from folium.features import DivIcon
import math

# Define the haversine function to calculate distance in kilometers
def haversine(coord1, coord2):
    lat1, lon1 = coord1
    lat2, lon2 = coord2

    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))

    r = 6371  # Earth radius in kilometers
    return c * r

# Example coordinates (replace with your actual data)
launch_site_coord = [28.5623, -80.5774]  # Launch site latitude and longitude
closest_point_coord = [28.5700, -80.5850]  # Closest city, railway, highway, etc.

# Calculate the distance
distance_km = haversine(launch_site_coord, closest_point_coord)

# Initialize the map centered between the two points
center_lat = (launch_site_coord[0] + closest_point_coord[0]) / 2
center_lon = (launch_site_coord[1] + closest_point_coord[1]) / 2
site_map = folium.Map(location=[center_lat, center_lon], zoom_start=12)

# Add marker for launch site
folium.Marker(
    location=launch_site_coord,
    popup='Launch Site',
    icon=folium.Icon(color='blue')
).add_to(site_map)

# Add marker for closest point with distance displayed using DivIcon
distance_marker = folium.Marker(
    location=closest_point_coord,
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12px; color: red;"><b>{distance_km:.2f} KM</b></div>'
    )
)
distance_marker.add_to(site_map)

# Draw a line between launch site and closest point
line = folium.PolyLine(locations=[launch_site_coord, closest_point_coord], color='green', weight=2)
line.add_to(site_map)

# Display the map
site_map
```
